### Aufgaben:
1. Die Logik kann der Funktion „entschluesseln“ (wird von uns bereitgestellt) entnommen werden.
1. Schaue Dir die Funktion „entschluesseln“ an und verstehe was sie macht. Es werden Funktionen benutzt, die du noch nicht kennst. Gerne hier das richtige Googeln schärfen.
1. Nutze die Funktion um die verschlüsselte Nachricht zu entschlüsseln.

#### Zu entziffernde Nachricht:
BPO YVN IDZHVIY RZDNN YVNN HZDI KVNNRJMO HDO ZDIN URZD NOVMOZO...

    
### Erweiterungsmöglichkeiten:
1. Erstellt eure eigene verschlüsselte Nachricht. Gerne auch mit einer anderen Verschiebung. Hinweis: Ihr könnt die „entschlüsseln“ Funktion auch nutzen / umbauen um eine Nachricht zu verschlüsseln.
1. Erweitere die „entschluessel“ Funktion, dass auch Zahlen verschoben werden.
    
### Gewünschte Lösung:


````python
def entschluesseln(verschluesselte_nachricht, verschiebung):
    enschluesselte_nachricht = ""

    for buchstabe in verschluesselte_nachricht:
        if buchstabe.isalpha():
            buchstabe_index = ord(buchstabe.upper()) - ord('A')
            verschobener_index = (buchstabe_index + verschiebung) % 26
            verschobener_buchstabe = chr(verschobener_index + ord('A'))
            enschluesselte_nachricht += verschobener_buchstabe
        else:
            enschluesselte_nachricht += buchstabe
    return enschluesselte_nachricht


# Hier deinen Code einfügen
````
